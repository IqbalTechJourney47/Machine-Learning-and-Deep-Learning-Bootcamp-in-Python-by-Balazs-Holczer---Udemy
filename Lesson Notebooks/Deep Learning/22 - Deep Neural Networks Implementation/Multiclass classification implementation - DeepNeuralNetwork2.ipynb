```python
import numpy as np
from keras.models import Sequential
# from keras.layers.core import Dense
# from tensorflow.keras.layers import Dense
from keras.layers import Dense
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.optimizers import Adam
# from tensorflow.keras.optimizers import Adam
```

```python
dataset = load_iris()
```

```python
features = dataset.data
print (features)
```

```python
y = dataset.target.reshape(-1, 1)
```

```python
# 1D array without reshape
print(dataset.target)
```

```python
# 2D array with reshape
print(y)
```

```python
encoder = OneHotEncoder()

# The OneHotEncoder by default returns a sparse matrix, and the Keras model have trouble working with this sparse matrix.
# To fix this issue, we need to convert the sparse matrix to a dense numpy array before using it for training.

# targets = encoder.fit_transform(y)
targets = encoder.fit_transform(y).toarray()  # Convert to dense numpy array
```

```python
encoder = OneHotEncoder(sparse=False)
targets = encoder.fit_transform(y)

# If we do not do something like this (sparse=False), then python is not going to transform the target variables into a
# one hot encoded vector as we desire
```

One hot encoding

It means we are going to represent data in 1-D array

```python
train_features, test_features, train_targets, test_targets = train_test_split(features, targets, test_size=0.2)
```

```python
model = Sequential()

# first parameter is output dimension
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(10, input_dim=10, activation='relu'))
model.add(Dense(10, input_dim=10, activation='relu'))
model.add(Dense(3, activation='softmax'))

# 3 output neurons
# model.add(Dense(3, activation='softmax'))
```

```python
# we can define the loss function MSE or negative log likelihood
# optimizer will find the right adjustments for the weights: SGD, Adagrad, ADAM ...

# optimizer = Adam(lr=0.005)   # lr = learning rate
learning_rate = 0.005
optimizer = Adam(learning_rate=learning_rate)

model.compile(loss = 'categorical_crossentropy',
             optimizer = optimizer,
             metrics = 'accuracy')
```

Mean Squared Error (MSE)

We use mean squared error when we are dealing with regression(so we have only one output feature)

Negative Log Likelihood

We use negative log likelihood function when dealing with classification


```python
print("train_features dtype:", train_features.dtype)
print("train_targets dtype:", train_targets.dtype)
```

```python
print("Shape of train_features:", train_features.shape)
print("Shape of train_targets:", train_targets.shape)
```

```python
model.fit(train_features, train_targets, epochs=1000, batch_size=20, verbose=2)

# batch_size=20
# this is mini batch
# it means that the underlying Adam Optimizer will update the model parameters such as W edge weights after
# considering 20 items in the training dataset.

# verbose=2
# it means that we would like to print out the accuracy, the loss function value in every single iteration
```

```python
results = model.evaluate(test_features, test_targets)
print('Accuracy on the test dataset: %.2f' % results[1])

# The result is going to have two parameters
# If we call index is zero, it is going to print out the loss function value
# But we are after the accuracy, this is why we get the second parameter
```
